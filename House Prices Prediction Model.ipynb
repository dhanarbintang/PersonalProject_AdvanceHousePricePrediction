```python
feature_list = ['LotArea','BedroomAbvGr','MoSold','YrSold','HalfBath','FullBath','MSZoning','OverallQual','OverallCond','Neighborhood','SalePrice']
df1 = train[feature_list]
display(df1.head())
print(df1.shape)
```

# Data Preparation & Cleaning

```python
display(df1.groupby("BedroomAbvGr")["BedroomAbvGr"].agg('count'))
```

```python
df2 = df1[~(df1.BedroomAbvGr>5) | (df1.BedroomAbvGr==0)]
display(df2)
print(df2.shape)
```

```python
df3 = df2.copy()
df3['price_per_sqft'] = df3['SalePrice']/df3['LotArea']
print(df3.shape)
df3
```

# Outlier Removal

```python
df3.price_per_sqft.describe()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby(['MSZoning','Neighborhood']):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df4 = remove_pps_outliers(df3)
display(df4)
df4.shape
```

# Model Building

```python
df5 = df4.drop('price_per_sqft', axis='columns')
df5
```

```python
dummies = pd.get_dummies(df5.MSZoning)
display(dummies)

dummies2 = pd.get_dummies(df5.Neighborhood)
display(dummies2)
```

```python
df6 = df5.drop(['MSZoning','Neighborhood'], axis='columns')
df6
```

```python
df7 = pd.concat([df6, dummies.drop('C (all)', axis='columns')], axis='columns')
df7
```

```python
df8 = pd.concat([df7, dummies2.drop('BrDale', axis='columns')], axis='columns')
df8
```

```python
from sklearn.model_selection import train_test_split

def split_data(df):
    X_train, X_val, y_train, y_val = train_test_split(df.loc[:, df.columns != 'SalePrice'], 
                                                      df['SalePrice'], test_size=0.15, random_state=2022)
    return X_train, X_val, y_train, y_val

X_train, X_val, y_train, y_val = split_data(df8)
display(X_train)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
from xgboost import XGBRegressor

model = XGBRegressor()
model.fit(X_train, y_train, verbose=False)
pred = model.predict(X_val)
```

```python
from sklearn.metrics import mean_squared_error

def rmsle(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))

print("RMSLE : " + str(rmsle(y_val, pred)))
```

```python
X_test = test[feature_list[:-1]]
X_test = X_test.drop(['MSZoning','Neighborhood'], axis='columns')
X_test = pd.concat([X_test, dummies.drop('C (all)', axis='columns')], axis='columns')
X_test = pd.concat([X_test, dummies2.drop('BrDale', axis='columns')], axis='columns')
X_test
```
